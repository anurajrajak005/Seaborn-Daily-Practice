```python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Initialize the Relational Graph
G = nx.Graph()

# 2. Add Entities (Nodes) and Relations (Edges)
# Example: Entity A relates to Entity B
G.add_edge("Bron", "Foundation")
G.add_edge("Bron", "Structure")
G.add_edge("Bron", "Network")

# 3. Plot the Relational Graph
plt.figure(figsize=(8, 6))

# Define layout (how nodes are distributed)
pos = nx.spring_layout(G, seed=42)

# Draw nodes and edges
nx.draw_networkx_nodes(G, pos, node_size=700, node_color="skyblue")
nx.draw_networkx_edges(G, pos, width=2, edge_color="gray")
nx.draw_networkx_labels(G, pos, font_size=12, font_family="sans-serif")

# Display the plot
plt.title("Relational Plot of Bron")
plt.axis("off")
plt.show()
```

```python
# Load a built-in example dataset
df = sns.load_dataset("penguins")

# 1. Histogram (Default)
sns.displot(data=df, x="flipper_length_mm", kind="hist")

# 2. Kernel Density Estimate (KDE) - smooth curve
sns.displot(data=df, x="flipper_length_mm", kind="kde")

# 3. Empirical Cumulative Distribution Function (ECDF)
sns.displot(data=df, x="flipper_length_mm", kind="ecdf")

plt.show()
```

```python
# Compute correlation matrix and plot
df = sns.load_dataset("tips")
matrix = df.corr(numeric_only=True)
sns.heatmap(matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
df = sns.load_dataset("tips")

# Create a basic point plot
sns.pointplot(data=df, x="day", y="total_bill")

# Show the plot
plt.show()
```
